```python
db_name = 'my_database.db'

# Check if the database file exists in the current directory
if not os.path.isfile(db_name):
    
    conn = sqlite3.connect(db_name)

    # conn.execute('''
    # CREATE TABLE IF NOT EXISTS users (
    #     id INTEGER PRIMARY KEY AUTOINCREMENT,
    #     name TEXT NOT NULL,
    #     age INTEGER
    # )
    # ''')
    print("Table 'users' created successfully.")    
    
    conn.close()
else:
    print(f"Database '{db_name}' already exists.")
```

```python
def sql_tool(input=""):
    db_path = "my_database.db"
    db = SQLDatabase.from_uri(f"sqlite:///{db_path}")
    toolkit = create_sql_agent(db=db, llm=llm, verbose = True)
    res = toolkit.run(input)
    return (res)

sql_prompt = """
    You are a SQL expert agent interacting with a my_database.db database. Your job is to write SQL queries to create or retrieve data based on user instructions.
"""

sql = Tool(
    name = "SQL_AGENT",
    func = sql_tool,
    description = sql_prompt
)
```
